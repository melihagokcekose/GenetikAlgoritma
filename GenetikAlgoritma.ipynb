```python
import random
#import os

class Room: #Öğretim görevlilerinin ofis bilgilerini tutan sınıf
    def __init__(self, number, seating_capacity):
        self.number = number
        self.seating_capacity = seating_capacity

    def get_number(self):
        return self.number

    def get_seating_capacity(self):
        return self.seating_capacity

class Instructor: #Öğretim görevlilerinin bilgilerini tutan sınıf
    def __init__(self, _id, name):
        self.id = _id
        self.name = name

    def get_id(self):
        return self.id

    def get_name(self):
        return self.name

class MeetingTime: #Ders saatlerini tutan sınıf
    def __init__(self, _id, time):
        self.id = _id
        self.time = time

    def get_id(self):
        return self.id

    def get_time(self):
        return self.time

class Course: #Kurs bilgilerini tutan sınıf
    def get_max_numb_of_students(self):
        return self.max_num_of_students
    def __init__(self, number, name, max_num_of_students, instructors):
        self.number = number
        self.name = name
        self.max_num_of_students = max_num_of_students
        self.instructors = instructors

    def get_number(self):
        return self.number

    def get_name(self):
        return self.name

    def get_max_num_of_students(self):
        return self.max_num_of_students

    def get_instructors(self):
        return self.instructors

class Department: #Bölüm bilgilerini tutan sınıf
    def __init__(self, name, courses):
        self.name = name
        self.courses = courses

    def get_name(self):
        return self.name

    def get_courses(self):
        return self.courses

class Class: #Derslik bilgilerini tutan sınıf
    def __init__(self, _id, dept, course):
        self.id = _id
        self.dept = dept
        self.course = course
        self.instructor = None
        self.meeting_time = None
        self.room = None

    def set_instructor(self, instructor):
        self.instructor = instructor

    def set_meeting_time(self, meeting_time):
        self.meeting_time = meeting_time

    def set_room(self, room):
        self.room = room

    def get_id(self):
        return self.id

    def get_dept(self):
        return self.dept

    def get_course(self):
        return self.course

    def get_instructor(self):
        return self.instructor

    def get_meeting_time(self):
        return self.meeting_time

    def get_room(self):
        return self.room

    def __str__(self):
        return f"[{self.dept.get_name()}, {self.course.get_number()}, {self.room.get_number()}, {self.instructor.get_id()}, {self.meeting_time.get_id()}]"

class Data:
    def __init__(self):
        self.rooms = []
        self.instructors = []
        self.courses = []
        self.depts = []
        self.meeting_times = []
        self.number_of_classes = 0
        self.initialize()

    def initialize(self):
        room1 = Room("R1", 20)
        room2 = Room("R2", 25)
        room3 = Room("R3", 35)
        self.rooms = [room1, room2, room3]

        meeting_time1 = MeetingTime("MT1", "PÇC 09:00 - 10:00")
        meeting_time2 = MeetingTime("MT2", "PÇC 10:00 - 11:00")
        meeting_time3 = MeetingTime("MT3", "SPH 09:00 - 10:30")
        meeting_time4 = MeetingTime("MT4", "SPH 10:30 - 12:00")
        self.meeting_times = [meeting_time1, meeting_time2, meeting_time3, meeting_time4]

        instructor1 = Instructor("I1", "Dr.Ebubekir Koç")
        instructor2 = Instructor("I2", "Prf.Dr.Burhannetin Can")
        instructor3 = Instructor("I3", "Prf.Dr.Fevzi Yılmaz")
        instructor4 = Instructor("I4", "Dr.Berna Kiraz")
        self.instructors = [instructor1, instructor2, instructor3, instructor4]

        course1 = Course("C1", "325K", 25, [instructor1, instructor2, instructor4])
        course2 = Course("C2", "319K", 35, [instructor4, instructor2, instructor3])
        course3 = Course("C3", "416K", 20, [instructor1, instructor2])
        course4 = Course("C4", "464K", 25, [instructor3, instructor4])
        course5 = Course("C5", "360C", 35, [instructor4])
        course6 = Course("C6", "303K", 35, [instructor1, instructor3])
        course7 = Course("C7", "303LK", 20, [instructor2, instructor4])
        self.courses = [course1, course2, course3, course4, course5, course6, course7]

        dept1 = Department("BLM", [course1, course4])
        dept2 = Department("EEM", [course2, course4, course5])
        dept3 = Department("BMD", [course6, course7])
        self.depts = [dept1, dept2, dept3]

        self.number_of_classes = sum(len(dept.get_courses()) for dept in self.depts)

    def get_rooms(self):
        return self.rooms

    def get_instructors(self):
        return self.instructors

    def get_courses(self):
        return self.courses

    def get_depts(self):
        return self.depts

    def get_meeting_times(self):
        return self.meeting_times

    def get_number_of_classes(self):
        return self.number_of_classes

class Population:
    def __init__(self, size, data):
        self.schedules = [Schedule(data).initialize() for _ in range(size)]

    def get_schedules(self):
        return self.schedules

    def sort_by_fitness(self):
        self.schedules.sort(key=lambda schedule: schedule.get_fitness(), reverse=True)
        return self

class GeneticAlgorithm:
    def __init__(self, data):
        self.data = data

    def evolve(self, population):
        return self.mutate_population(self.crossover_population(population))

    def crossover_population(self, population):
        crossover_population = Population(len(population.get_schedules()), self.data)
        for x in range(MainSchedule.NUMB_OF_ELITE_SCHEDULES):
            crossover_population.get_schedules()[x] = population.get_schedules()[x]

        for x in range(MainSchedule.NUMB_OF_ELITE_SCHEDULES, len(population.get_schedules())):
            if MainSchedule.MUTATION_RATE > random.random():
                schedule1 = self.select_tournament_population(population).sort_by_fitness().get_schedules()[0]
                schedule2 = self.select_tournament_population(population).sort_by_fitness().get_schedules()[0]
                crossover_population.get_schedules()[x] = self.crossover_schedule(schedule1, schedule2)
            else:
                crossover_population.get_schedules()[x] = population.get_schedules()[x]

        return crossover_population

    def one_point_crossover_schedule(self, schedule1, schedule2):
        one_point_crossover_schedule = Schedule(self.data).initialize()

        r1 = len(schedule1.get_classes())
        index = int(random.random() * r1)

        for x in range(len(one_point_crossover_schedule.get_classes())):
            if x > index:
                one_point_crossover_schedule.get_classes()[x] = schedule1.get_classes()[x]
            else:
                one_point_crossover_schedule.get_classes()[x] = schedule2.get_classes()[x]

        return one_point_crossover_schedule

    def two_point_crossover_schedule(self, schedule1, schedule2):
        two_point_crossover_schedule = Schedule(self.data).initialize()

        first_index = int(schedule1.get_classes() * random.random())
        second_index = int(schedule2.get_classes() * random.random())

        for x in range(len(two_point_crossover_schedule.get_classes())):
            if second_index > first_index:
                two_point_crossover_schedule.get_classes()[x] = schedule1.get_classes()[x]
            else:
                two_point_crossover_schedule.get_classes()[x] = schedule2.get_classes()[x]

        return two_point_crossover_schedule

    def crossover_schedule(self, schedule1, schedule2):
        crossover_schedule = Schedule(self.data).initialize()

        for x in range(len(crossover_schedule.get_classes())):
            if random.random() > 0.5:
                crossover_schedule.get_classes()[x] = schedule1.get_classes()[x]
            else:
                crossover_schedule.get_classes()[x] = schedule2.get_classes()[x]

        return crossover_schedule

    def mutate_population(self, population):
        mutate_population = Population(len(population.get_schedules()), self.data)
        schedules = mutate_population.get_schedules()

        for x in range(MainSchedule.NUMB_OF_ELITE_SCHEDULES):
            schedules[x] = population.get_schedules()[x]

        for x in range(MainSchedule.NUMB_OF_ELITE_SCHEDULES, len(population.get_schedules())):
            schedules[x] = self.mutate_schedule(population.get_schedules()[x])

        return mutate_population

    def mutate_schedule(self, mutate_schedule):
        schedule = Schedule(self.data).initialize()

        for x in range(len(mutate_schedule.get_classes())):
            if MainSchedule.MUTATION_RATE > random.random():
                mutate_schedule.get_classes()[x] = schedule.get_classes()[x]

        return mutate_schedule

    def select_tournament_population(self, population):
        tournament_population = Population(MainSchedule.TOURNAMENT_SELECTION_SIZE, self.data)
        for x in range(MainSchedule.TOURNAMENT_SELECTION_SIZE):
            tournament_population.get_schedules()[x] = population.get_schedules()[int(random.random() * len(population.get_schedules()))]

        return tournament_population


class RandomNumbers:
    _instance = None

    def __init__(self):
        self._random = random.Random()

    @classmethod
    def get_instance(cls):
        if not cls._instance:
            cls._instance = cls()
        return cls._instance

    def set_seed(self, seed=None):
        if seed is not None:
            self._random = random.Random(seed)
        else:
            self._random = random.Random()

    def get_random_number(self):
        return self._random.random()


class Schedule: #Program bilgilerini tutan sınıf
    def __init__(self, data):
        self.data = data
        self.classes = []
        self.is_fitness_changed = True
        self.fitness = -1
        self.num_of_conflicts = 0
        self.class_numb = 0

    def get_data(self):
        return self.data

    def initialize(self):
        for dept in self.data.get_depts():
            for course in dept.get_courses():
                new_class = Class(self.class_numb, dept, course)
                new_class.set_meeting_time(random.choice(self.data.get_meeting_times()))
                new_class.set_room(random.choice(self.data.get_rooms()))
                new_class.set_instructor(random.choice(course.get_instructors()))
                self.classes.append(new_class)
                self.class_numb += 1
        return self

    def get_num_of_conflicts(self):
        return self.num_of_conflicts

    def get_classes(self):
        self.is_fitness_changed = True
        return self.classes

    def get_fitness(self):
        if self.is_fitness_changed:
            self.fitness = self.calculate_fitness()
            self.is_fitness_changed = False
        return self.fitness

    def calculate_fitness(self):
        self.num_of_conflicts = 0
        for x in self.classes:
            if x.get_room().get_seating_capacity() < x.get_course().get_max_numb_of_students():
                self.num_of_conflicts += 1
            for y in self.classes[self.classes.index(x):]:
                if x.get_meeting_time() == y.get_meeting_time() and x.get_id() != y.get_id():
                    if x.get_room() == y.get_room():
                        self.num_of_conflicts += 1
                    if x.get_instructor() == y.get_instructor():
                        self.num_of_conflicts += 1

        return 1 / (self.num_of_conflicts + 1)

    def __str__(self):
        return ', '.join(map(str, self.classes))



class MainSchedule:

    POPULATION_SIZE = 10
    MUTATION_RATE = 0.1
    CROSSOVER_RATE = 0.9
    TOURNAMENT_SELECTION_SIZE = 3
    NUMB_OF_ELITE_SCHEDULES = 1
    SCHEDULE_FILE = "input_data/schedule.txt"

    def __init__(self):
        self.schedule_number = 0
        self.class_number = 0
        self.data = Data()

    def main(self):
        generation_number = 0
        self.print_available_data()

        print("> Generation #", generation_number)
        print("  Schedule # |                                  ", end="")
        print("Classes [dept,class,room,instructor,meeting-time]       ", end="")
        print("                                               |Fitness | Conflicts ")
        print(
            "-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

        genetic_algorithm = GeneticAlgorithm(self.data)
        population = Population(MainSchedule.POPULATION_SIZE, self.data).sort_by_fitness()
        for schedule in population.get_schedules():
            print(f"     {self.schedule_number}      | {schedule} | {schedule.get_fitness():.5f} | {schedule.get_num_of_conflicts()}")
            self.schedule_number += 1

        self.print_schedule_as_table(population.get_schedules()[0], generation_number)
        self.class_number = 1

        while population.get_schedules()[0].get_fitness() != 1.0:
            generation_number += 1
            print("> Generation #", generation_number)
            print("  Schedule # |                                  ", end="")
            print("Classes [dept,class,room,instructor,meeting-time]", end="")
            print("                                                                                  | Fitness | Conflicts ")
            print(
                "-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

            population = genetic_algorithm.evolve(population).sort_by_fitness()
            for schedule in population.get_schedules():
                print(f"     {self.schedule_number}      | {schedule} | {schedule.get_fitness():.5f} | {schedule.get_num_of_conflicts()}")
                self.schedule_number += 1

            self.print_schedule_as_table(population.get_schedules()[0], generation_number)
            self.class_number = 1

    def print_schedule_as_table(self, schedule, generation):
        classes = schedule.get_classes()
        print("\n                     ")
        print("Class # | Dept   | Course (number, max # of students ) | Room (Capacity) |       Instructor (Id)    |   Meeting Time (Id)  ")
        print("                     ")
        print("---------------------------------------------------------------------------------------------------------------------------------")

        for x in range(len(classes)):
            major_index = self.data.get_depts().index(classes[x].get_dept())
            courses_index = self.data.get_courses().index(classes[x].get_course())
            room_index = self.data.get_rooms().index(classes[x].get_room())
            instructors_index = self.data.get_instructors().index(classes[x].get_instructor())
            meetin_time_index = self.data.get_meeting_times().index(classes[x].get_meeting_time())

            print("                     ", end="")
            print(f"{str(self.class_number).zfill(2)}    | ", end="")
            print(f"{self.data.get_depts()[major_index].get_name():^4} | ", end="")
            print(f"{self.data.get_courses()[courses_index].get_name()} ({self.data.get_courses()[courses_index].get_number()}, {classes[x].get_course().get_max_num_of_students()})           |", end="")
            print(f"{self.data.get_rooms()[room_index].get_number()} ({classes[x].get_room().get_seating_capacity()})     | ", end="")
            print(f"{self.data.get_instructors()[instructors_index].get_name()} ( {self.data.get_instructors()[instructors_index].get_id()})      | ", end="")
            print(f"{self.data.get_meeting_times()[meetin_time_index].get_time()} ({self.data.get_meeting_times()[meetin_time_index].get_id()})")
            self.class_number += 1

        if schedule.get_fitness() == 1:
            print(f"> Solution Found in {generation + 1} generations")

        print("-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------")

    def print_available_data(self):
        print("Available Departments ==>")
        for dept in self.data.get_depts():
            print(f"name: {dept.get_name()}, courses: {dept.get_courses()}")

        print("\nAvailable Courses ==>")
        for course in self.data.get_courses():
            print(f"course: {course.get_number()}, max # of students: {course.get_max_num_of_students()}, instructors: {course.get_instructors()}")

        print("\nAvailable Rooms ==>")
        for room in self.data.get_rooms():
            print(f"room: {room.get_number()}, max seating capacity: {room.get_seating_capacity()}")

        print("\nAvailable Instructors ==>")
        for instructor in self.data.get_instructors():
            print(f"id: {instructor.get_id()}, name: {instructor.get_name()}")

        print("\nAvailable Meeting Times ==>")
        for meeting_time in self.data.get_meeting_times():
            print(f"id: {meeting_time.get_id()}, Meeting Time: {meeting_time.get_time()}")

        print("-----------------------------------------------------------------------------------")
        print("-----------------------------------------------------------------------------------")


if __name__ == "__main__":
    main_schedule = MainSchedule()
    main_schedule.main()
```
